```python
# Matrix Multiplication with User Input

# Input size of first matrix
r1 = int(input("Enter number of rows of Matrix A: "))
c1 = int(input("Enter number of columns of Matrix A: "))

# Input size of second matrix
r2 = int(input("Enter number of rows of Matrix B: "))
c2 = int(input("Enter number of columns of Matrix B: "))

# Check if multiplication is possible
if c1 != r2:
    print("Matrix multiplication not possible!")
else:
    # Input Matrix A
    print("Enter elements of Matrix A:")
    A = []
    for i in range(r1):
        row = []
        for j in range(c1):
            row.append(int(input(f"A[{i}][{j}]: ")))
        A.append(row)

    # Input Matrix B
    print("Enter elements of Matrix B:")
    B = []
    for i in range(r2):
        row = []
        for j in range(c2):
            row.append(int(input(f"B[{i}][{j}]: ")))
        B.append(row)

    # Initialize Result Matrix with zeros
    result = []
    for i in range(r1):
        result.append([0] * c2)

    # Matrix Multiplication
    for i in range(r1):
        for j in range(c2):
            for k in range(c1):
                result[i][j] += A[i][k] * B[k][j]

    # Print Result
    print("Product of matrices:")
    for row in result:
        print(row)
```
